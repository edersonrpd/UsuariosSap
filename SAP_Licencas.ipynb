````python
import pandas as pd
import numpy as np
import datetime
import matplotlib
````

````python
licenca_sp_df = pd.read_excel('/content/drive/MyDrive/SAP/Administração de licenças_sp.xlsx')
log_acesso_sp_df = pd.read_excel('/content/drive/MyDrive/SAP/Log de acesso_sp.xlsx')
````

````python
condicao = [
    (licenca_sp_df['Profissional'] == '✔'),
    (licenca_sp_df['Limited Logistics'] == '✔'),
    (licenca_sp_df['Indirect Access'] == '✔'),
    ((licenca_sp_df['Profissional'] != '') & (licenca_sp_df['Limited Logistics'] != '') & (licenca_sp_df['Indirect Access'] != ''))
  
   ]

valores = ['Profissional', 'Limited Logistics', 'Indirect Access', 'Sem Licença']

licenca_sp_df['Licença'] = np.select(condicao, valores)
````

````python
sp_df = pd.merge(licenca_sp_df, log_acesso_sp_df[['Código do usuário','Nome do usuário', 'Super-usuário', 'Último logon']], on='Código do usuário', how='left')
sp_df = sp_df[['Código do usuário', 'Nome do usuário', 'Licença', 'Super-usuário', 'Último logon']]
sp_df.rename(columns={'Último logon' : 'Último logon SP'}, inplace=True)
````



---



#Combinando tabelas

````python
usuario_sp = sp_df[['Código do usuário', 'Nome do usuário', 'Licença', 'Super-usuário']]
usuario_go = go_df[['Código do usuário', 'Nome do usuário', 'Licença', 'Super-usuário']]
usuario_mt = mt_df[['Código do usuário', 'Nome do usuário', 'Licença', 'Super-usuário']]
usuario_rj = rj_df[['Código do usuário', 'Nome do usuário', 'Licença', 'Super-usuário']]

usuarios = pd.concat([usuario_sp, usuario_go, usuario_mt, usuario_rj])
usuarios = usuarios.drop_duplicates()
usuarios = pd.merge(usuarios, sp_df[['Código do usuário', 'Último logon SP']], on=['Código do usuário'], how='left' )
usuarios = pd.merge(usuarios, go_df[['Código do usuário', 'Último logon GO']], on=['Código do usuário'], how='left' )
usuarios = pd.merge(usuarios, mt_df[['Código do usuário', 'Último logon MT']], on=['Código do usuário'], how='left' )
usuarios = pd.merge(usuarios, rj_df[['Código do usuário', 'Último logon RJ']], on=['Código do usuário'], how='left' )
````

````python
usuarios[['Último logon SP', 'Último logon GO', 'Último logon MT', 'Último logon RJ']] = usuarios[['Último logon SP', 'Último logon GO', 'Último logon MT', 'Último logon RJ']].fillna('01/01/1900 00:00:01')
usuarios
````

````python

colunas = ['Último logon SP','Último logon GO','Último logon MT', 'Último logon RJ']

for coluna in colunas:
  usuarios[coluna] = pd.to_datetime(usuarios[coluna], format='%d/%m/%Y %H:%M:%S')
````

````python
usuarios = usuarios.sort_values(by='Nome do usuário')

datatypes_per_column = {
    "Último logon SP" : "datetime64[ns]",
    "Último logon GO" : "datetime64[ns]",
    "Último logon MT" : "datetime64[ns]",
    "Último logon RJ" : "datetime64[ns]",
}
usuarios = usuarios.astype(datatypes_per_column)

````

## Criando coluna com o ultimo logon

````python
def max_value(row):
  return max(row['Último logon SP'], row['Último logon GO'], row['Último logon MT'], row['Último logon RJ'])

usuarios['Último logon'] = usuarios.apply(max_value, axis=1)
````

````python
usuarios = usuarios[['Código do usuário', 'Nome do usuário', 'Licença', 'Super-usuário', 'Último logon',
       'Último logon SP', 'Último logon GO', 'Último logon MT',
       'Último logon RJ']]

usuarios = usuarios.drop_duplicates()
````

````python
columns_to_format = ['Último logon','Último logon SP', 'Último logon GO', 'Último logon MT', 'Último logon RJ']

for column in columns_to_format:
    usuarios[column] = usuarios[column].dt.strftime('%d/%m/%Y %H:%M:%S')
````

````python
usuarios = usuarios[usuarios['Licença'] != 'Sem Licença']
usuarios
````
